```python
import random

num_simulations = 1000000

count_A = 0
count_B = 0
count_A_and_B = 0

for _ in range(num_simulations):
    A = random.random()
    B = random.random()

    if A < 0.5:
        count_A += 1
    if B < 0.3:
        count_B += 1
    if A < 0.5 and B < 0.3:
        count_A_and_B += 1

P_A = count_A / num_simulations
P_B = count_B / num_simulations
P_A_and_B = count_A_and_B / num_simulations
P_B_given_A = P_A_and_B / P_A if P_A > 0 else 0

print(f"P(A) = {P_A}")
print(f"P(B) = {P_B}")
print(f"P(A ∩ B) = {P_A_and_B}")
print(f"P(B|A) = {P_B_given_A}")
```

```python
import random
from collections import Counter

num_simulations = 1000000

count_A = 0
count_B = 0

urn = ['red'] * 6 + ['blue'] * 4 + ['green'] * 6

for _ in range(num_simulations):
    draw = random.sample(urn, 3)

    if 'red' in draw:
        count_A += 1

    if len(set(draw)) == 1:
        count_B += 1

P_A = count_A / num_simulations
P_B = count_B / num_simulations

print(f"P(A) = {P_A}")
print(f"P(B) = {P_B}")
```

```python
import random
import matplotlib.pyplot as plt

def roll_dice(n):
    return [random.randint(1, 6) for _ in range(n)]

def plot_histogram(data, title):
    plt.hist(data, bins=range(1, 8), align='left', rwidth=0.8, density=True, color='turquoise', edgecolor='black')
    plt.title(title)
    plt.xlabel('Die Face')
    plt.ylabel('Relative Frequency')
    plt.xticks(range(1, 7))
    plt.show()

data_200 = roll_dice(200)
plot_histogram(data_200, 'Histogram for 200 Rolls')

data_2000 = roll_dice(2000)
plot_histogram(data_2000, 'Histogram for 2000 Rolls')
```

```python
import random
import matplotlib.pyplot as plt
import numpy as np

def simulate_dice_rolls(n):
    return [random.randint(1, 6) + random.randint(1, 6) + random.randint(1, 6) for _ in range(n)]

def theoretical_probabilities():
    counts = [0]*19
    for i in range(1, 7):
        for j in range(1, 7):
            for k in range(1, 7):
                counts[i+j+k] += 1
    return [count/216 for count in counts]

def plot_histogram(data, theoretical, title):
    plt.hist(data, bins=range(3, 20), align='left', rwidth=0.8, density=True, color='blue', edgecolor='black', alpha=0.5, label='Simulation')
    plt.bar(range(3, 19), theoretical[3:], color='red', alpha=0.5, label='Theoretical')
    plt.title(title)
    plt.xlabel('Sum of Dice')
    plt.ylabel('Relative Frequency')
    plt.legend()
    plt.show()

for n in [500, 1000, 2000, 5000]:
    data = simulate_dice_rolls(n)
    theoretical = theoretical_probabilities()
    plot_histogram(data, theoretical, f'Histogram for {n} Rolls')

best_number = np.argmax(theoretical[3:]) + 3
print(f'The best number to bet on is {best_number}.')

print(f'The theoretical probability of {best_number} occurring is {theoretical[best_number]:.3f}.')
```

```python
import random

num_simulations = 1000000

count_success = 0

for _ in range(num_simulations):
    birthdays = [random.randint(1, 12) for _ in range(5)]

    if len(set(birthdays)) == 4:
        count_success += 1

P_simulation = count_success / num_simulations

print(f"Simulation Probability = {P_simulation}")

P_theoretical = (12 * 11 * 10 * 9 * (5 * 4 / 2)) / (12**5)

print(f"Theoretical Probability = {P_theoretical}")
```

```python
import random

num_simulations = 1000000

count_success_a = 0
count_success_b = 0

for _ in range(num_simulations):
    roll1 = random.randint(1, 6)
    roll2 = random.randint(1, 6)

    if roll1 == 4 and roll1 + roll2 >= 7:
        count_success_a += 1

    if roll2 % 2 == 0 and roll1 + roll2 >= 7:
        count_success_b += 1

P_simulation_a = count_success_a / num_simulations
P_simulation_b = count_success_b / num_simulations

print(f"Simulation Probability a) = {P_simulation_a}")
print(f"Simulation Probability b) = {P_simulation_b}")

P_theoretical_a = sum(1 for i in range(1, 7) if 4 + i >= 7) / 6
P_theoretical_b = sum(1 for i in range(1, 7) for j in range(1, 7) if j % 2 == 0 and i + j >= 7) / 36

print(f"Theoretical Probability a) = {P_theoretical_a}")
print(f"Theoretical Probability b) = {P_theoretical_b}")
```
