```python
#a)

mu = 199
sigma = 3

simulated_data = norm.rvs(mu, sigma, 1000)

mean_value = np.mean(simulated_data)

print(f"Mittlere abgefullte Menge Tee in einer Packung: {mean_value}")
```

```python
#b)
mu = 199
sigma = 3

simulated_data = norm.rvs(mu, sigma, 1000)

probab_less_195 = np.mean(simulated_data < 195)

prob_let_195_get_198 = np.mean((195 <= simulated_data) & (simulated_data <= 198))

prob_more_than_195 = np.mean(simulated_data > 195)

probab_less_195_theoretical = norm.cdf(195, mu, sigma)
prob_let_195_get_198_theoretical = norm.cdf(198, mu, sigma) - norm.cdf(195, mu, sigma)
prob_more_than_195_theoretical = 1 - norm.cdf(195, mu, sigma)

print(f"Simulation for prob less than 195: {probab_less_195}")
print(f"Simulation for prob between 195 and 198: {prob_let_195_get_198}")
print(f"Simulation for prob more than 195: {prob_more_than_195}")
print(f"Theoretical simulation for prob less than 195: {probab_less_195_theoretical}")
print(f"Theoretical simulation for prob between 195 and 198: {prob_let_195_get_198_theoretical}")
print(f"Theoretical simulation for prob more than 195: {prob_more_than_195_theoretical}")
```

```python
mu = 199
sigma = 3

simulated_data = norm.rvs(mu, sigma, 1000)

hist_count, bin_edges = np.histogram(simulated_data, bins = 16)

print("Anzahl der Daten in jeder Klasse: ")
for i, count in enumerate(hist_count):
    print(f"Klasse {i + 1}: {count} Daten")
    
plt.hist(simulated_data, bins=16, density=True, edgecolor = "black", label="relative Hfg.")

x = np.linspace(min(simulated_data), max(simulated_data), 100)
pdf = norm.pdf(x, mu, sigma)
plt.plot(x, pdf, label = "Dichtefkt")
plt.xlabel("Abgefullte Menge Tree")
plt.ylabel("Relative Hfg")
plt.legend()
plt.show()
```

```python
num_simulations = 10000
bet_amount = 6
num_people = 6

def simulate_birthday_bet():
    birthdays = np.random.randint(1, 366, size=num_people)
    if len(birthdays) != len(set(birthdays)):
        return bet_amount
    else:
        return -bet_amount
    
simulated_results = np.array([simulate_birthday_bet() for _ in range(num_simulations)])

average_profit_loss = np.mean(simulated_results)

win_probability = np.mean(simulated_results > 0)

print("Durchschnittlicher Gewinn oder Verlust:", average_profit_loss)
print("Wahrscheinlichkeit zu gewinnen:", win_probability)
```

```python
num_simulations = 10000

simulated_results = np.random.choice([0,1, 2], size = num_simulations)

print("Simulationsergebnisse:")
for num_black_balls in [0, 1, 2]:
    count = np.sum(simulated_results == num_black_balls)
    print(f"Anzahl der gezogenen schwarzen Kugeln = {num_black_balls} : {count} Simulationen")

print("\nTheoretische Wahrscheinlichkeiten:")
print(f"P(X=0) = {2/5}")
print(f"P(X=1) = {4/15}")
print(f"P(X=2) = {1/6}")
```

```python
points_distribution = {0: -5, 1: -5, 2: -5, 3: 30}

simulated_points = np.array([points_distribution[num_black_balls] for num_black_balls in simulated_results])

average_points = np.mean(simulated_points)

print("\nSimulationsergebnisse fr Punkte:")
print(f"Durchschnittliche Punktezahl: {average_points}")

theoretical_points = sum(prob * points_distribution[num_black_balls] for num_black_balls, prob in zip([0, 1, 2], [2/5, 4/15, 1/6]))

print(f"Theoretische Punktezahl: {theoretical_points}")
```

```python
num_points = 1000

points = np.random.uniform(low = -1, high = 1, size=(num_points, 3))

distances = np.linalg.norm(points, axis=1)

mean_distance = np.mean(distances)
variance_distance = np.var(distances)

print("Erwartungswert der Distanz:", mean_distance)
print("Varianz der Distanz:", variance_distance)
```
