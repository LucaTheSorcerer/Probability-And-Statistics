```python
import random
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm, binom
```

# Exercise 1

 Der Vektor D entha ̈lt 800 Daten, welche die Werte der ZG
       4    5   6   7    8   9  10
  X ∼ 0.05 0.1 0.1 0.35 0.2 0.1 0.1
sind.
- a) Man scha ̈tze den Erwartungswert E(X), die Varianz V(X), die Wahrscheinlichkeiten P(X ≤ 7), P(X > 4).
Hinweis: numpy.mean, numpy.var
- b) Man gebe an die theoretischen Werte fu ̈r E(X),V(X),P(X ≤ 7),P(X > 4).
Hinweis: ErwartungswertE(X)=SumxiP(X=xi);VarianzV(X)=E(X2)−E2(X). i∈I
- c) Anhand der simulierten Daten konstruiere man die Histogramme der relativen, bzw. absoluten H ̈aufigkeiten.

```python
X = [4, 5, 6, 7, 8, 9, 10]
P=[0.05,0.1,0.1,0.35,0.2, 0.1, 0.1] 
D=random.choices(X, weights=P, k=500)
print(D)

D_np = np.array(D)
E_X = np.mean(D_np)
V_X = np.var(D_np)

P_X_leq_7 = np.sum(D_np <= 7) / len(D_np)
P_X_gt_4 = np.sum(D_np > 4) / len(D_np)

print("E(X) = ", E_X)
print("V(X) = ", V_X)
print("P(X <= 7) = ", P_X_leq_7)
print("P(X > 4) = ", P_X_gt_4)


#Compute the theoretical mean and variance of X

E_X_theo = np.sum(np.array(X) * np.array(P))
E_X2_theo = np.sum(np.array(X)**2 * np.array(P))
V_X = E_X2_theo - E_X_theo**2

P_X_leq7_theo = np.sum(np.array(X) <= 7)
P_X_gt4_theo = np.sum(np.array(X) > 4)

print("E(X) = ", E_X_theo)
print("V(X) = ", V_X)
print("P(X <= 7) = ", P_X_leq7_theo)
print("P(X > 4) = ", P_X_gt4_theo)


abs_freq = np.bincount(D_np)

rel_freq = abs_freq / len(D_np)


#plot histogram for absolute frequency

plt.bar(range(len(abs_freq)), abs_freq, color='b', alpha=0.5)
plt.xticks(range(len(abs_freq)), range(len(abs_freq)))
plt.xlabel('x')
plt.ylabel('Absolute Frequency')
plt.show()



#plot histogram for relative frequency

plt.bar(range(len(rel_freq)), rel_freq, color='b', alpha=0.5)
plt.xticks(range(len(rel_freq)), range(len(rel_freq)))
plt.xlabel('x')
plt.ylabel('Relative Frequency')
plt.show()

```
