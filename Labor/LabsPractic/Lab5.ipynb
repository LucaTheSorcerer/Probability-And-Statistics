# Exercise 1

Teepackungen, die von einer bestimmten Firma abgefu ̈llt werden, sollten mit jeweils 200 g Inhalt abgefu ̈llt werden. Die abgefu ̈llte Menge Tee X in einer Packung ist normal verteilt X ∼ N(μ,σ2); die dafu ̈r zusta ̈ndige Abfu ̈llmaschine hat eine Standardabweichung von σ = 3 g und ist auf einen Erwartungswert μ = 199 g eingestellt.
- a) Anhand 1000 simulierten Daten, welche ist im Mittel die abgefu ̈llte Menge Tee in einer Packung? Hinweis: Man benutze scipy.stats.norm.rvs(μ,σ,1000) fu ̈r die Generierung von Daten und danach numpy.mean.

- b) Mit welcher Wahrscheinlichkeit werden in einer Packung weniger als 195 g Tee abgefu ̈llt? Mit welcher WahrscheinlichkeitwerdenineinerPackungzwischen195gund198gTeeabgefu ̈llt? Mitwelcher Wahrscheinlichkeit werden in einer Packung mehr als 195 g Tee abgefu ̈llt? Man vergleiche das Ergeb- nis mit den theoretischen Wahrscheinlichkeiten.
Hinweis: Fu ̈r die theoretischen Wahrscheinlichkeiten benutze man scipy.stats.norm.cdf(x, μ, σ).
Es gilt: P(X ≤ x) = F(x) und P(X ∈ [a,b]) = F(b) − F(a) fu ̈r die stetige ZG X ∼ N(μ,σ2), F ist die Verteilungsfunktion (norm.cdf) von X.

- c) Die generierten Daten der Stichprobe sollen in 16 Klassen (Intervallen) eingeteilt. Man za ̈hle mit Hfg, Klasse=numpy.histogram(Daten, bins=16) und gebe an (mit print ) wie viele Daten in jeder Klasse sind. Man zeichne das entsprechende Histogramm der relativen H ̈aufigkeiten mit matplotlib.pyplot.hist(Daten,bins=16,density=True,edgecolor="black",label="rel. Hfg.") Auf demselben Bild zeichne man auch die Dichtefunktion der N(μ,σ2) Verteilung (μ = 199,σ = 3). Hinweis: Man benutze scipy.stats.norm.pdf(x, μ, σ)) und plot.

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
from scipy.stats import expon, uniform
import random
```

```python
mu = 199
sigma = 3
simulations = 1000

simulated_data = norm.rvs(loc=mu, scale=sigma, size=simulations)

average_amount = np.mean(simulated_data)
print("Average amount: ", average_amount)
```

```python
probability_less_than_195g = norm.cdf(195, mu, sigma)
probability_between_195g_and_198g = norm.cdf(198, mu, sigma) - norm.cdf(195, mu, sigma)
probability_more_than_195g = 1 - norm.cdf(195, mu, sigma)

print("Probability less than 195g: ", probability_less_than_195g)
print("Probability between 195g and 198g: ", probability_between_195g_and_198g)
print("Probability more than 195g: ", probability_more_than_195g)
```

```python
hist, bins = np.histogram(simulated_data, bins=16)

for i, count in enumerate(hist):
    print(f"Class {i+1}: {count} data points")

plt.hist(simulated_data, bins=16, density=True, edgecolor="black", label="Simulated data")
plt.xlabel("Tea amount (g)")
plt.ylabel("Probability")
plt.legend()
plt.show()
```

# Exercise 3

Jedesmal, wenn Professor X eine Gruppe von 6 Personen trifft, wettet er 6 e, dass mindestens zwei von diesen 6 Personen im gleichen Monat Geburtstag haben. Anhand Simulationen scha ̈tze man: den durchschnittlichen Gewinn oder Verlust bei dieser Wette, bzw. die Wahrscheinlichkeit p, mit welcher Professor X eine Wette gewinnt.
Hinweis:
.
Es gilt p = P (”mindestens 2 Personen von 6 haben in demselben Monat Geburtstag”).

```python
simulations = 10000

total_profit_wins = 0
total_wins = 0

for _ in range(simulations):

    birthdays = [random.randint(1, 12) for _ in range(6)]

    if len(birthdays) != len(set(birthdays)):
        total_profit_wins += 6
        total_wins += 1
    else:
        total_profit_wins -= 6


average_proft_win = total_profit_wins / simulations
win_probability = total_wins / simulations

print(f"Average profit win: {average_proft_win:.2f}")
print(f"Win probability: {win_probability:.2f}")
```

# Exercise 5

Man simuliere mit Hilfe von scipy.stats.uniform.rvs 1000 zuf ̈allige Punkte aus dem Quader [−1, 1] × [−1, 1] × [−1, 1] ⊂ R3.
Sei D die ZG welche Distanz dieser Punkte zum Ursprung (0,0,0) darstellt. Man sch ̈atze a) den Er- wartungswert (numpy.mean) und die Varianz von D (numpy.var).

```python
simulations = 1000
random_points = uniform.rvs(loc=-1, scale=2, size=(simulations, 3))

# print(random_points)

distances = np.linalg.norm(random_points, axis=1)

#Compute the mean distances

mean_distance = np.mean(distances)

#Compute the variance

variance_distance = np.var(distances)

print(f"Estimated mean distance: {mean_distance}")
print(f"Estiamted variance of distance: {variance_distance}")
```
