Die unabhängigen Zufallsgrößen X und Y haben foldende Wahrscheinlichkeitsverteilung
       -1   0    1   2
X,Y ~ (025 0.25 025 025)
 - a) Man generiere 2000 zufällige Daten für die Zufallsgröße X + Y. Man gebe die erhaltenen Werte und die entsprechenden
relativen Häufigkeiten an, zB. mit print (f" X+y=... hat relative Hfg.=...").
 - (b) Man zeichne das Histogramm der relativen Häufigkeiten von den 2000 simulierten Werten für X + Y
 - (c) Man schätze P(X + Y < 1).
 - (b) Man gebe die theoretische Wahrscheinlichkeitsverteilung von X + Y an, d.h. alle möglichen Werte die X + Y annehmen
kann und die entsprechenden Wahrscheinlichkeiten, zB. mit print (f" P (X+Y=. . .) =.. ").
 - (c) Man gebe die theoretischen Wert an für P(X + Y < 1).

```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

# Define the values and probabilities
values = np.array([-1, 0, 1, 2])
probs = np.array([0.25, 0.25, 0.25, 0.25])

# Generate 2000 random data for X + Y
X = np.random.choice(values, p=probs, size=2000)
Y = np.random.choice(values, p=probs, size=2000)
Z = X + Y

# Calculate the relative frequencies
counts = Counter(Z)
relative_freqs = {k: v / 2000 for k, v in counts.items()}

# Print the obtained values and their relative frequencies
for k, v in relative_freqs.items():
    print(f"X + Y = {k} has relative frequency = {v}")

# Draw the histogram of the relative frequencies
plt.hist(Z, bins=np.arange(Z.min(), Z.max()+2)-0.5, edgecolor='black', density=True)
plt.xlabel('X + Y')
plt.ylabel('Relative Frequency')
plt.title('Histogram of Relative Frequencies')
plt.show()

# Estimate P(X + Y < 1)
prob_less_than_1 = np.mean(Z < 1)
print(f"Estimated P(X + Y < 1) = {prob_less_than_1}")

# Theoretical probability distribution of X + Y
theoretical_probs = {i+j: probs[i]*probs[j] for i in values for j in values}
print("Theoretical probability distribution of X + Y:")
for k, v in theoretical_probs.items():
    print(f"P(X + Y = {k}) = {v}")

# Theoretical value for P(X + Y < 1)
theoretical_prob_less_than_1 = sum(v for k, v in theoretical_probs.items() if k < 1)
print(f"Theoretical P(X + Y < 1) = {theoretical_prob_less_than_1}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

# Define the values and their probabilities
values = [-1, 0, 1, 2]
probabilities = [0.25, 0.25, 0.25, 0.25]

# (a) Generate 2000 random data for the random variable X + Y
X = np.random.choice(values, 2000, p=probabilities)
Y = np.random.choice(values, 2000, p=probabilities)
Z = X + Y

# Print the obtained values and their relative frequencies
counter = Counter(Z)
for value, count in counter.items():
    print(f"X+Y={value} has relative frequency={count/2000}")

# (b) Draw the histogram of the relative frequencies of the 2000 simulated values for X + Y
plt.hist(Z, bins=np.arange(min(Z), max(Z) + 2) - 0.5, edgecolor='black', density=True)
plt.xlabel('X + Y')
plt.ylabel('Relative Frequency')
plt.show()

# (c) Estimate P(X + Y < 1)
p_less_than_1 = sum([1 for value in Z if value < 1]) / 2000
print(f"P(X + Y < 1) = {p_less_than_1}")

# (d) Give the theoretical probability distribution of X + Y
# Since X and Y are independent and identically distributed, the possible values of X + Y range from -2 to 4
# The probabilities can be calculated by convolving the probability distribution with itself
theoretical_probabilities = np.convolve(probabilities, probabilities)
for i, p in enumerate(theoretical_probabilities):
    print(f"P(X+Y={i-2}) = {p}")

# (e) Give the theoretical value for P(X + Y < 1)
# This is the sum of the theoretical probabilities for X + Y = -2, -1, 0
theoretical_p_less_than_1 = sum(theoretical_probabilities[:3])
print(f"Theoretical P(X + Y < 1) = {theoretical_p_less_than_1}")
```
