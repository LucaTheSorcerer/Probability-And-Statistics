Hinweis: Fu ̈r die theoretischen Wahrscheinlichkeiten benutze man scipy.stats.norm.cdf(x, μ, σ).
Es gilt: P(X ≤ x) = F(x) und P(X ∈ [a,b]) = F(b) − F(a) fu ̈r die stetige ZG X ∼ N(μ,σ2), F ist die Verteilungsfunktion (norm.cdf) von X.

```python
#a)

mu = 199
sigma = 3

simulated_data = norm.rvs(mu, sigma, 1000)

mean_value = np.mean(simulated_data)

print(f"Mittlere abgefullte Menge Tee in einer Packung: {mean_value}")
```

```python
#b)
mu = 199
sigma = 3

simulated_data = norm.rvs(mu, sigma, 1000)

probab_less_195 = np.mean(simulated_data < 195)

prob_let_195_get_198 = np.mean((195 <= simulated_data) & (simulated_data <= 198))

prob_more_than_195 = np.mean(simulated_data > 195)

probab_less_195_theoretical = norm.cdf(195, mu, sigma)
prob_let_195_get_198_theoretical = norm.cdf(198, mu, sigma) - norm.cdf(195, mu, sigma)
prob_more_than_195_theoretical = 1 - norm.cdf(195, mu, sigma)

print(f"Simulation for prob less than 195: {probab_less_195}")
print(f"Simulation for prob between 195 and 198: {prob_let_195_get_198}")
print(f"Simulation for prob more than 195: {prob_more_than_195}")
print(f"Theoretical simulation for prob less than 195: {probab_less_195_theoretical}")
print(f"Theoretical simulation for prob between 195 and 198: {prob_let_195_get_198_theoretical}")
print(f"Theoretical simulation for prob more than 195: {prob_more_than_195_theoretical}")
```

```python
mu = 199
sigma = 3

simulated_data = norm.rvs(mu, sigma, 1000)

hist_count, bin_edges = np.histogram(simulated_data, bins = 16)

print("Anzahl der Daten in jeder Klasse: ")
for i, count in enumerate(hist_count):
    print(f"Klasse {i + 1}: {count} Daten")
    
plt.hist(simulated_data, bins=16, density=True, edgecolor = "black", label="relative Hfg.")

x = np.linspace(min(simulated_data), max(simulated_data), 100)
pdf = norm.pdf(x, mu, sigma)
plt.plot(x, pdf, label = "Dichtefkt")
plt.xlabel("Abgefullte Menge Tree")
plt.ylabel("Relative Hfg")
plt.legend()
plt.show()
```

# Exercise 3

Jedesmal, wenn Professor X eine Gruppe von 6 Personen trifft, wettet er 6 e, dass mindestens zwei von diesen 6 Personen im gleichen Monat Geburtstag haben. Anhand Simulationen scha ̈tze man: den durchschnittlichen Gewinn oder Verlust bei dieser Wette, bzw. die Wahrscheinlichkeit p, mit welcher Professor X eine Wette gewinnt.
Hinweis:
                                            -6    6
W (Gewinn, bzw. Verlust, bei einer Wette) ∼ 1 − p p


```python
import numpy as np

# Number of simulations
N = 100000

# Number of people in a group
group_size = 6

# Bet amount
bet_amount = 6

# Function to simulate a group meeting and return the result of the bet
def simulate_bet():
    # Generate random birthdays for the group
    birthdays = np.random.randint(1, 13, group_size)
    # Check if at least two people have the same birthday
    return bet_amount if len(birthdays) != len(set(birthdays)) else -bet_amount

# Simulate N bets
results = [simulate_bet() for _ in range(N)]

# Calculate the average gain or loss
average_result = np.mean(results)
print(f"Average gain or loss: {average_result} euros")

# Calculate the probability of winning a bet
p = len([result for result in results if result > 0]) / N
print(f"Probability of winning a bet: {p}")
```

# Exercies 4

In einer Urne sind 4 weiße und 6 schwarze Kugeln. Ein Spieler zieht nacheinander eine Kugel ohne Zuru ̈cklegen. Das Spiel ist aus, wenn er eine weisse Kugel zieht oder wenn er dreimal gezogen hat. Die Zufallsvariable X zeigt die Anzahl der gezogenen schwarzen Kugeln.
- a) Welches ist die (theoretische) Wahrscheinlichkeitsverteilung von X und simuliere zuf ̈allige Werte fu ̈r X. 
- b) Der Spieler erha ̈lt 30 Punkte, wenn er drei schwarze Kugeln gezogen hat. Er erh ̈alt 25 Punkte, wenn er zwei schwarze Kugeln zieht. In allen anderen Fa ̈llen verliert er 5 Punkte. Anhand Simulationen scha ̈tze man die mittlere Punktezahl des Spielers. Man vergleiche das Ergebnis mit dem theoretischen Ergebnis. 
    Hinweise: 
    - a) Die Zufallsvariable X ist die Anzahl der gezogenen schwarzen Kugeln. Die (theoretische) Wahrscheinlichkeitsverteilung von X ist:
    P(X=0)=P(ersteKugelistweiß)= 4/10
    P (X = 1) = P (erste Kugel schwarz, zweite Kugel weiß) = ...
    P (X = 2) = P (erste und zweite Kugel schwarz, dritte Kugel weiß) = ... P (X = 3) = P (erste und zweite und dritte Kugel schwarz) = ...

    - b) Die Zufallsvariable Y zeigt die Punktezahl des Spielers an; die (theoretische) Wahrscheinlichkeitsverteilung von Y ist:
    P(Y =30)=...
    P(Y =25)=...
    P(Y =−5)=...
    =⇒ der theoretische Erwartungswert von Y ist E(Y ) = ... (Punkte).

```python
N = 10000

def simulate_game():
    urn = ['white'] * 4 + [ 'black'] * 6
    draws = 0
    for _ in range(3):
        ball = np.random.choice(urn)
        urn.remove(ball)
        if ball == 'black':
            draws += 1
        else:
            break

    return draws

#simulate n games
results = [simulate_game() for _ in range(N)]

#Calculate the (theoretical) probability of winning a game
prob_distributions = {i: results.count(i) / N for i in range(4)}
print(f"Probability of winning a game: {prob_distributions}")

def calculate_score(x):
    if x == 3:
        return 30
    elif x == 2:
        return 25
    else:
        return - 5
    
average_score = np.mean([calculate_score(x) for x in results])
print(f"Avergae score of the player: {average_score}")
```

# Exercise 5

Man simuliere mit Hilfe von scipy.stats.uniform.rvs 1000 zuf ̈allige Punkte aus dem Quader [−1, 1] × [−1, 1] × [−1, 1] ⊂ R3.
Sei D die ZG welche Distanz dieser Punkte zum Ursprung (0,0,0) darstellt. Man schatze 
- a) den Erwartungswert (numpy.mean) und die Varianz von D (numpy.var).

```python
N = 1000

points = uniform.rvs(loc=-1, scale=2, size=(N, 3))

distances = np.sqrt(np.sum(points**2, axis=1))

expected_value = np.mean(distances)
print(f"Expected value of the distance: {expected_value}")

variance = np.var(distances)
print(f"Variance of the distance: {variance}")
```
