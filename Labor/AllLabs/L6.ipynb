```python
'''
Man gebe an die theoretischen Werte fu ̈r E(X),V(X),P(X ≤ 7),P(X > 4).
Hinweis: ErwartungswertE(X)=SumxiP(X=xi);VarianzV(X)=E(X2)−E2(X). i∈I
'''

X = np.array([4, 5, 6, 7, 8, 9, 10])
P = np.array([0.05, 0.1, 0.1, 0.35, 0.2, 0.1, 0.1])

E_X = np.sum(X * P)

E_X2 = np.sum((X**2) * P)
V_X = E_X2 - E_X**2

P_X_leq_7 = np.sum(P[X <= 7])
P_X_gt_4 = np.sum(P[X > 4])

print(f"E(X): {E_X}")
print(f"V(X): {V_X}")
print(f"P(X ≤ 7): {P_X_leq_7}")
print(f"P(X > 4): {P_X_gt_4}")

```

# Exercise 3

Ein Computer ist mit zwei Druckern verbunden: D1 and D2. Der Computer schickt einen Druck- auftrag an D1 mit Wahrscheinlichkeit 0.4, beziehungsweise an D2 mit Wahrscheinlichkeit 0.6. D1 druckt einen A2 Poster in T1 Sekunden, wobei T1 ∼ Exp(1/5), D2 druckt einen A2 Poster in T2 Sekunden, wobei T2 ∼ Unif[4,6]. Ein Druckauftrag fur einen A2 Poster wird abgeschickt.
- a) Man schatze die Wahrscheinlichkeit, dass das Drucken mehr als 5 Sekunden dauert?
- b) Man schatze den Erwartungswert und die Standardabeweichung fur die Druckzeit (in Sekunden) des Posters.
Hinweis: scipy.stats.uniform(loc=...,scale=...,size=...) scipy.stats.expon(loc=0,scale=...,size=...),
numpy.mean, numpy.std

```python
prob_D1 = 0.4
prob_D2 = 0.6

printing_time_D1 = expon(scale = 51)
printing_time_D2 = uniform(loc = 4,scale = 2)

N = 10000
print_times = np.zeros(N)

for i in range(N):
    chosen_printer = np.random.choice(['D1', 'D2'], p=[prob_D1, prob_D2])
    
    if chosen_printer == 'D1':
        print_times[i] = printing_time_D1.rvs()
    else:
        print_times[i] = printing_time_D2.rvs()

probability_more_than_5_seconds = np.mean(print_times > 5)

mean_printing_time = np.mean(print_times)
std_printing_time = np.std(print_times)

print(f"a) Estimated probability of printing taking more than 5 seconds: {probability_more_than_5_seconds:.4f}")
print(f"b) Estimated mean printing time: {mean_printing_time:.2f} seconds")
print(f"   Estimated standard deviation of printing time: {std_printing_time:.2f} seconds")
```

 # Exercise 4

 Sei die Gleichung zweiten Grades x**2+Bx+C = 0, wobei B,C ∼ Unif[−1,1] unabhangige ZG sind. Man schatze:
- a) die Wahrscheinlichkeit, dass beide Wurzeln der Gleichung reell sind;
- b) die Wahrscheinlichkeit, dass beide Wurzeln der Gleichung positiv sind;
- c) den Erwartungswert und die Varianz der Summe der beiden Wurzeln.

```python
N = 10000

B_values = uniform(loc =-1, scale=2).rvs(N)
C_values = uniform(loc =-1, scale=2).rvs(N)

discr = B_values**2 - 4*C_values

prob_real_roots = np.mean(discr >= 0)

positive_roots_cond = (discr >= 0) & (-B_values > 0)
prob_positive_roots = np.mean(positive_roots_cond)

roots_sum = -B_values / 2
mean_roots_sum = np.mean(roots_sum)
var_roots_sum = np.var(roots_sum)

print(f"Prob that both roots are real: {prob_real_roots:.4f}")
print(f"Prob that both roots are positive: {prob_positive_roots:.4f}")
print(f"Estimation of mean of sum of roots: {mean_roots_sum:.4f}")
print(f"Estimation of variance of sum of roots: {var_roots_sum:.4f}")
```

# Exercise 5

In einer Urne sind 20 rote Kugeln, 15 blaue Kugeln, 5 grune Kugeln und 10 schwarze Kugeln. Man simuliere N (= 200, 1000, . . .) Ziehungen mit Zuru ̈cklegen und zeige (print) die relative H ̈aufigkeit an mit welcher jede Farbe auftaucht. Man vergleiche die theoretischen Resultate mit den Ergebnissen aus den Simulationen. Man gebe die Ergebnisse der ersten 10 Ziehungen an!

```python
import numpy as np


urn = {'red': 20, 'blue': 15, 'green': 5, 'black': 10}

Ns = [200, 1000, 5000]

for N in Ns:
    draws = np.random.choice(list(urn.keys()), size=N, replace=True)

    print(f"\nResults for {N} simulations:")
    for i in range(10):
        color = draws[i]
        relative_frequency = draws[:i+1].tolist().count(color) / (i+1)
        print(f"Draw {i+1}: {color.capitalize()} - Relative Frequency: {relative_frequency:.4f}")

    theoretical_probabilities = {color: count / sum(urn.values()) for color, count in urn.items()}

    simulated_relative_frequencies = {color: draws.tolist().count(color) / N for color in urn.keys()}

    print("\nTheoretical Probabilities:")
    for color, probability in theoretical_probabilities.items():
        print(f"{color.capitalize()}: {probability:.4f}")

    print("\nSimulated Relative Frequencies:")
    for color, frequency in simulated_relative_frequencies.items():
        print(f"{color.capitalize()}: {frequency:.4f}")
```

# Exercise 6

Eine Urne entha ̈lt 10 Kugeln mit der Ziffer 0, 20 Kugeln mit der Ziffer 1, 20 Kugeln mit der Ziffer 2. Aus der Urne werden 3 Kugeln ohne Zuru ̈cklegen gezogen. X sei das Produkt der 3 erhaltenen Zahlen. Man sch ̈atze anhand Simulationen den Erwartungswert und die Varianz von X! Man erstelle anhand Simulationen das Histogramm der absoluten H ̈aufigkeiten fu ̈r die Werte von X! In ein zweites Bild zeichne man ein zweites Histogramm mit den (theoretischen) Wahrscheinlichkeiten der ZG X.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the urn
urn = [0]*10 + [1]*20 + [2]*20

# Simulate drawing 3 balls without replacement 10000 times
simulations = [np.prod(np.random.choice(urn, size=3, replace=False)) for _ in range(10000)]

# Calculate the expected value and variance of X
expected_value = np.mean(simulations)
variance = np.var(simulations)

print(f"Expected value: {expected_value}")
print(f"Variance: {variance}")

# Create a histogram of the absolute frequencies for the values of X
plt.hist(simulations, bins=range(9), alpha=0.5, color='g', edgecolor='black')
plt.title('Histogram of absolute frequencies for X')
plt.xlabel('X')
plt.ylabel('Frequency')
plt.show()

# Create a histogram with the (theoretical) probabilities of the random variable X
unique, counts = np.unique(simulations, return_counts=True)
probabilities = counts / len(simulations)
plt.bar(unique, probabilities, alpha=0.5, color='b', edgecolor='black')
plt.title('Histogram of probabilities for X')
plt.xlabel('X')
plt.ylabel('Probability')
plt.show()
```

# Exercise 7

In einer Urne sind 3 blaue, 3 rote und 4 weiße Kugeln. Ein Spieler zieht nacheinander ohne Zurucklegen 3 Kugeln. Der Spieler erhalt 5 e, wenn alle drei gezogenen Kugeln dieselbe Farbe haben. Er erhalt 2 e, wenn die drei Kugeln unterschiedliche Farben aufweisen. Bei allen anderen Fallen muss der Spieler 1 e bezahlen. Wie viel gewinnt oder verliert im Mittel der Spieler pro Spiel? Man vergleiche das theoretische Resultat mit den Ergebnissen der Simulationen.

```python
urn = ['blue'] * 3 + ['red']*3 + ['white']*4

def payout(draw):
    if len(set(draw)) == 1:
        return 5
    elif len(set(draw)) == 3:
        return 2
    else:
        return -1
    
simulations = [payout(np.random.choice(urn, size=3, replace=False)) for _ in range(1000)]

average_gain_loss = np.mean(simulations)

print(f"Average gain/loss: {average_gain_loss}")
```

```python
# Define the urn with 3 blue, 3 red, and 4 white balls
urn = ['blue'] * 3 + ['red'] * 3 + ['white'] * 4

# Function to simulate one game
def simulate_game():
    # Shuffle the urn
    random.shuffle(urn)
    
    # Draw 3 balls without replacement
    drawn_balls = urn[:3]
    
    # Check for the outcome and calculate winnings/losses
    if len(set(drawn_balls)) == 1:  # All balls have the same color
        return 5
    elif len(set(drawn_balls)) == 3:  # All balls have different colors
        return 2
    else:
        return -1

# Number of simulations
num_simulations = 10000

# Simulate and store the game results
results = [simulate_game() for _ in range(num_simulations)]

# Calculate the average winnings/losses
average_result = sum(results) / num_simulations

# Print the estimated average winnings/losses
print(f'Average Winnings/Losses per Game: €{average_result:.2f}')
```

# Exercise 8

- 1) Man stelle die Dichtefunktion bzw. die Verteilungsfunktion fu ̈r 
    - a) X ∼ Unif[−2,2];
    - b) X ∼ Exp(2);
grafisch dar auf den Intervallen: fu ̈r a) [-3,3]; fu ̈r c) [0,4]!
2) Man sch ̈atze in jedem Fall anhand Simulationen P (1 < X < 1.5). Man vergleiche den gesch ̈atzen Wert mit dem theoretischen Wert indem man die Python Befehle fu ̈r die Verteilungsfunktionen benutzt!
3) Man sch ̈atze in jedem Fall anhand Simulationen den Erwartungswert E(X) und die Varianz V (X). Hinweis: scipy.stats.uniform , scipy.stats.expon



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform, expon

# Define the distributions
distributions = [uniform(loc=-2, scale=4), expon(scale=1/2)]
names = ['Uniform[-2,2]', 'Exponential(2)']
intervals = [np.linspace(-3, 3, 1000), np.linspace(0, 4, 1000)]

# Plot the density function and the distribution function for each distribution
for distribution, name, interval in zip(distributions, names, intervals):
    plt.figure(figsize=(12, 6))
    plt.plot(interval, distribution.pdf(interval), label='Density function')
    plt.plot(interval, distribution.cdf(interval), label='Distribution function')
    plt.title(name)
    plt.legend()
    plt.show()

# Estimate P(1 < X < 1.5) using simulations and compare with the theoretical value
for distribution, name in zip(distributions, names):
    simulations = distribution.rvs(size=10000)
    estimated_prob = np.mean((1 < simulations) & (simulations < 1.5))
    theoretical_prob = distribution.cdf(1.5) - distribution.cdf(1)
    print(f"{name}: Estimated P(1 < X < 1.5) = {estimated_prob}, Theoretical P(1 < X < 1.5) = {theoretical_prob}")

# Estimate the expected value E(X) and the variance V(X) using simulations
for distribution, name in zip(distributions, names):
    simulations = distribution.rvs(size=10000)
    estimated_mean = np.mean(simulations)
    estimated_variance = np.var(simulations)
    print(f"{name}: Estimated E(X) = {estimated_mean}, Estimated V(X) = {estimated_variance}")
```
